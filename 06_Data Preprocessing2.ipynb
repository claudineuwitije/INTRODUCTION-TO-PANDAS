# Renaming a column

```python
df2=df1.rename({'First Name':'First_Name','Start Date':'Start_Date', 'Last Login Time':'Last_Login_Time',
       'Bonus %':'Bonus', 'Senior Management':'Senior_Management'}, axis=1) # Renaming columns
df2
```

```python
df1.rename({'First Name':'First_Name','Start Date':'Start_Date', 'Last Login Time':'Last_Login_Time',
       'Bonus %':'Bonus', 'Senior Management':'Senior_Management'}, axis=1, inplace=True) # Renaming columns
df1
```

# Adding a column

```python
df2['Bonus2']=df2.Bonus.round() # adding a column to DataFrame( In this case a column"Bonus2" has been added)
```

# Adding a row using  concat

```python
df3=df2.iloc[:2]
df3
```

```python
df4=pd.concat((df2,df3),axis=0) # merging the df2 and df3 data frame using concat
df4
```

```python
df4=pd.concat((df2,df3),axis=0,ignore_index=True) # Here you are telling pandas to ignore the indexes of df3 
df4                                               # and continue from the last index of df2                
```

# remove a rows  or columns

```python
df5=df4.drop(['First_Name'], axis=1) # removing the column "First_Name" using **drop**
df5
```

```python
df6=df4.drop(range(5),axis=0) # removing the column "First_Name" Deleting the first 5 rows
df6
```
