# Reading Data from File 

```python
import pandas as pd
```

```python
df1=pd.read_csv('Dataset/data_01.csv')  # Data_01 is read without any issue
df1
```

```python
df2=pd.read_csv('Dataset/data_02.csv')
df2           # Data_02 does not have a header, so the first line was used as the header of the DataFrame
```

```python
df2=pd.read_csv('Dataset/data_02.csv',header=None)  # setting the header to "None"
df2
```

```python
df2=pd.read_csv('Dataset/data_02.csv',header=None, names=['F_Name','L_Name','Course','Score'])  # labelling the header
df2
```

```python
df3=pd.read_csv('Dataset/data_03.csv')  
df3  # Data_03 is read as a single column because the original file has separators between variables
```

```python
len(df3.columns) # dataframe has only one column
```

```python
df3=pd.read_csv('Dataset/data_03.csv', sep='|')  # use sep ="type the separate in dataset"
df3
```

```python
df4=pd.read_csv('Dataset/data_04.csv') 
df4 # data_04 is read as a single column, and the first three lines are comments and not part of the DataFrame
```

```python
df4=pd.read_csv('Dataset/data_04.csv',comment='#',sep='\t')
df4  # comment='#' helps Pandas to understand that lines starting with # are treated as comments and should be ignored.
```
