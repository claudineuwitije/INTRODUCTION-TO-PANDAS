# DATA PROCESSING - Data Visualization

- Creating simple plots 
    - Line plots
    - Scatter plots
    - Histograms
    - Bar charts

```python
import pandas as pd
```

# Line plot 

- is used to find out the overall trend of how a particular feature changes over time.

```python
df1=pd.read_csv('INTRODUCTION-TO-PANDAS/Dataset/shampoo.csv')
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1.dtypes
```

```python
df1.Date=pd.to_datetime(df1.Date) # converting date into dateTime
```

```python
df1.dtypes
```

```python
df1.plot(x='Date',y='Sales')
```

```python
# Customizing the labels
df1.plot(x='Date',y='Sales',xlabel='Time',ylabel='Sales Numbers', title='Sales Numbers between 2000  and 2003',
        color='Green',linewidth=3,linestyle='--',marker='^',markersize=6,figsize=(8,6))
```

# Scatter plot 

- is used to represent relationship between two variables 

```python
df2=pd.read_csv('INTRODUCTION-TO-PANDAS/Dataset/IMDB-Movie-Data.csv')
```

```python
df2.head()
```

```python
df2.info()
```

```python
df2.plot(x='Rating',y='Metascore',kind='scatter',s=15,color='purple',marker='o')
```

# Histogram plot 

- is used to visualize how numerical value is distributed  accross dateset

```python
df2.plot(y='Runtime (Minutes)', kind='hist', bins=20, color='orange',figsize=(5,3))
```

# Bar chat

- is used to visualize the relative differences of values in features

```python
population=pd.Series(data=[300000000,126000000,37000000,212000000,83000000],index=['USA','Japan','Canada','Brazil','German'])
```

```python
population
```

```python
population.plot(kind='bar')
```

```python
population.plot(kind='pie')
```
